```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from logistic_regression import LogisticRegression

X = np.arange(1, 21).reshape(-1, 1)
# X = (X - X.mean()) / X.std()

y = np.array([0] * 10 + [1] * 10)

model = LogisticRegression(learning_rate=0.1, n_iterations=1000)
model.fit(X, y)

probs = model.predict_proba(X)

sns.scatterplot(x=X.flatten(), y=y, color='blue', label='Actual', s=20)
sns.lineplot(x=X.flatten(), y=probs, color='red', label='Sigmoid Curve')

plt.axhline(0.5, linestyle='--', color='gray', label='Decision Threshold')
plt.xlabel("X")
plt.ylabel("Class / Probability")
plt.title("Logistic Regression: Actual Labels vs Predicted Probability")
plt.legend()
plt.grid(True)

plt.savefig("logistic_regression_demo.png")

plt.show()
```
